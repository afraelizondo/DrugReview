```python
#Import libraries
import pandas as pd
import os 
```

```python
#Create DataFrame from the CSV
drug_df = pd.read_csv('drug_review_test.csv')
drug_df.head()
```

```python
#Keep relevant columns
drug_review_df = drug_df [['patient_id', 'drugName', 'condition', 'rating']]
drug_review_df.head()
```

```python
#Rename columns 
drugs_review_df = drug_review_df.rename(columns={
   "patient_id": "patient_id",
    "drugName": "drug_name",
    "condition": "med_condition", 
    "rating": "rating"
})

drugs_review_df.head()
```

```python
#Export as CSV(Removing the index column)
drugs_review_csv = drugs_review_df.to_csv('DrugReview.csv', index=False)
```
